### Economic Factors vs. Sales

```python
import pandas as pd
walmart = pd.read_csv('walmart_Sales.csv')
walmart.head()
```

```python
# how do fuel price, CPI, temperature, and unemployment affect sales performance?
FCTU_affect_sales = walmart.groupby(['Fuel_Price', 'CPI', 'Temperature', 'Unemployment'])['Weekly_Sales'].mean()
FCTU_affect_sales = FCTU_affect_sales.reset_index()
FCTU_affect_sales

```

```python
# Are there correlations between economics indicators and sales drops/spikes?
walmart['Date'] = pd.to_datetime(walmart['Date'], format='%d-%m-%Y')
corr_eco_sales = walmart.corr().Weekly_Sales
print(corr_eco_sales)
```
