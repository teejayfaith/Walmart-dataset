```python
import pandas as pd
```

```python
walmart_visual = pd.read_csv('walmart_Sales.csv')
walmart_visual.head()
```

```python
import matplotlib.pyplot as plt
```

### AVERAGE SALES PER MONTH

```python
# average sales per month
avg_monthly_sales = walmart_visual.groupby('Month')['Weekly_Sales'].mean()

# plot average sales per month
avg_monthly_sales.plot(kind='bar', figsize=(10, 6))
plt.title('Average Sales per Month')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.show()
```

WEEKLY SALES TREND PER YEAR

```python
# sales trends over time
walmart_visual['Date'] = pd.to_datetime(walmart_visual['Date'], format='%d-%m-%Y')
walmart_visual['Month'] = walmart_visual['Date'].dt.month_name()
walmart_visual['Year'] = walmart_visual['Date'].dt.year


weeklySales_trend = walmart_visual.groupby('Date')['Weekly_Sales'].sum()
plt.figure(figsize=(12,6))
weeklySales_trend.plot()
plt.title('Weekly Sales Trends Per Year')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.show()
# walmart_visual.head()
```
